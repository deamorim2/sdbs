```python
import duckdb

# Conectar ao DuckDB
con = duckdb.connect()

# Instalar e Carregar Extensão Espacial

con.sql('install spatial')
con.sql('load spatial')

# Caminhos do arquivo
csv_path = '/content/drive/My Drive/duckdb/open_buildings_v3_polygons_ne_110m_BRA.csv.gz'
parquet_path = '/content/drive/My Drive/duckdb/open_buildings_v3_polygons_ne_110m_BRA.parquet'

# Executar o comando COPY no DuckDB
con.execute(f"""
    COPY (
        SELECT * EXCLUDE geometry, ST_GeomFromText(geometry)
AS geometry FROM read_csv_auto('{csv_path}')
    ) TO '{parquet_path}' (FORMAT 'PARQUET');
""")

print(f"Arquivo convertido e salvo em: {parquet_path}")
```
